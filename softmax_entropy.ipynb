```python
import numpy as np
from scipy.stats import entropy
from scipy.special import softmax
import pandas as pd
```

```python
import json
data = []
guid= []
gold = []
dataset = "spam"
epoch = 4
with open(f'dy_log/{dataset}/roberta-base/training_dynamics/dynamics_epoch_{epoch}.jsonl') as f:
    for line in f:
        guid.append(json.loads(line)['guid'])
        data.append(json.loads(line)[f'logits_epoch_{epoch}'])
        gold.append(json.loads(line)['gold'])

# print(data[:5])
```

```python
en = []
sm = []
for line in data:
    print(line)
    sm.append(softmax(line))
    en.append(entropy(softmax(line)))

df = pd.DataFrame()
df['guid'] = guid
df['gold'] = gold
df['sm'] = sm
df['en'] = en
df.to_csv(f'dy_log/{dataset}/roberta-base/training_dynamics/final_{epoch}')
# np.savetxt(f"dy_log/{dataset}/roberta-base/training_dynamics/softmax_{epoch}_1", df)
# np.savetxt(f"dy_log/{dataset}/roberta-base/training_dynamics/entropy_{epoch}", en, delimiter=" ")
```
