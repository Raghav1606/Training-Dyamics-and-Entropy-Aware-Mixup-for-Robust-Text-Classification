```python
import numpy as np
from scipy.stats import entropy
from scipy.special import softmax
```

```python
import json
data = []
epoch = 4
with open(f'dy_log/imdb/roberta-base/training_dynamics/dynamics_epoch_{epoch}.jsonl') as f:
    for line in f:
        data.append(json.loads(line)[f'logits_epoch_{epoch}'])

# print(data[:5])
```

```python
en = []
sm = []
for line in data:
    print(line)
    sm.append(softmax(line))
    en.append(entropy(softmax(line)))

np.savetxt(f"dy_log/imdb/roberta-base/training_dynamics/softmax_{epoch}", sm, delimiter=" ")
np.savetxt(f"dy_log/imdb/roberta-base/training_dynamics/entropy_{epoch}", en, delimiter=" ")
```
